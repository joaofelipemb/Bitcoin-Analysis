## `Bitcoin` as the `'Digital Gold'`
 

Since Bitcoin became famous in 2017, when its price went from $1,000 to $19,000, many investors justify their investment decision in the crypto by stating that bitcoin is the new 'digital gold', meaning that it could be used as 'safe haven' protect the portfolio in times of volatilty and uncertainty in the stock market.

According to these investors, bitcoin is NOT correlated with the stock market, therefore, it would not suffer from risks of recession, high interest rates and other economic factors that might bring the stock market down.

This study has the purpose of testing if bitcoin can be used as a 'safe haven' asset by measuring its volatility and correlation against other assets. Also, test if the recent upward movement in bitcoin prices were actually generated by investors FOMO effect, instead of actual fundamentals.

```python
# Libraries
from datetime import datetime, date
import pandas as pd
import pandas_datareader as web
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
from math import sqrt
```

```python
# Importing price data

# bitcoin price data
btc_usd = web.DataReader('BTC-USD', data_source='yahoo', start='2017-03-23', end='2022-06-08')['Adj Close'].to_frame().rename(columns={'Adj Close': 'BTCUSD'})

# SP500 price data
sp500 = web.DataReader('^GSPC', data_source='yahoo', start='2017-03-23', end='2022-06-08')['Adj Close'].to_frame().rename(columns={'Adj Close': 'SP500'})

# NASDAQ 100 price data
nasdaq = web.DataReader('^IXIC', data_source='yahoo', start='2017-03-23', end='2022-06-08')['Adj Close'].to_frame().rename(columns={'Adj Close': 'NASDAQ100'})

# Gold futures price data
gold_usd = web.DataReader('GC=F', data_source='yahoo', start='2017-03-23', end='2022-06-08')['Adj Close'].to_frame().rename(columns={'Adj Close': 'GOLD'})

# 10Y Treasure Bond Yields
treasure_yields = web.DataReader('^TNX', data_source='yahoo', start='2017-03-23', end='2022-06-08')['Adj Close'].to_frame().rename(columns={'Adj Close': '10Y Treasury Bonds'})
```

```python
prices_df = pd.concat([btc_usd, sp500, nasdaq, gold_usd, treasure_yields], axis=1).dropna()
```

## `Year-to-date Performance Overview`

Now that we have our data available, let's go back to the `definition` of a sefe heven asset.

<b>'A safe haven is a type of investment that is expected to retain or increase in value during times of market turbulance'</b> -- Investopedia.


Now that we defined a safe haven asset, let's see if that holds true for bitcoin and compare it with gold. 

2022 Has been a difficult year for risk assets, if bitcoin is in fact the 'digital gold' it should be performing well during this market tumoil


```python
# Assets performance in 2022 so far.

ytd_performance = ((prices_df.loc['2022-05-13'] /prices_df.loc['2022-01-03']) - 1).to_frame().rename(columns={0: 'YTD Performance'})
```

```python
ytd_performance * 100
```

As we can see, bitcoin <b>fails</b> on our first test in hold its value during market turmoil. It is actually <b> the worst performing asset in our comparison.</b>

That is very different from gold, which is negotiating below its latest peak, but it is holding its value so far.

Here we have our first signs that bitcoin (and other cryptocurrencies) might behave just like any other risk asset, such as stocks.

## `Correlation Analysis`

Besides suffering during economic downturns, risk assets tend to have a `negative correlation` with interest rates. If bitcoin is actually a safe haven, it should a positive relationship with interest rates or no relationship at all. 

Let's check


```python
# Calculating the correlation among asset classes
prices_df.loc['2022-01-03':, :].corr()
```

As we can see, <b>bitcoin and stocks present a negative relationship with interest rates</b>. On top of that, stocks and Bitcoin have a <b> strong positive relationship</b>, supporting our thesis that the cryptocurrency is treated as a risky asset instead of a safe haven asset. 

On the other hand, gold shows a weak correlation against all asset classes. This shows that its price is less likely to be affected by the price movement from the assets in this analysis

Even though bitcoins's negative relationship with interest rates is relatively small, when compared to stocks, the cryptocurrency might face troubles during 2022, a year that the FED must adopt a hawkish monetary policy to counter record high inflation.

We can have a visual representation of this correlation by plotting their performance during this year

```python
ytd_data = prices_df.loc['2022-01-03':, :].pct_change()
```

```python
plt.figure(figsize=(18, 10))
plt.plot(ytd_data.index, (ytd_data['BTCUSD'] + 1).cumprod() - 1, label='BTCUSD')
plt.plot(ytd_data.index, (ytd_data['SP500'] + 1).cumprod() - 1, label='SP500')
plt.plot(ytd_data.index, (ytd_data['NASDAQ100'] + 1).cumprod() - 1, label='NASDAQ100')
plt.plot(ytd_data.index, (ytd_data['GOLD'] + 1).cumprod() - 1, label='GOLD')
plt.plot(ytd_data.index, (ytd_data['10Y Treasury Bonds'] + 1).cumprod() - 1, label='10Y Treasury')
plt.vlines(ytd_data.index.to_frame()['Date'].loc['2022-05-04'], ymax=1, ymin=-.5, colors='r', label='Fed Interest 1st Hike')
plt.legend()
None
```

Here we can see how poorly risky assets are performing while the interest rate futures are soaring. 

However, it important to remember that the Interest Rate Futures upward movement <b> IS NOT</b> the actual Federal Reserve interest rate hike. It just shows the investors' <b>exepectation</b> for the interest rate in the future. It still has the power to move markets though.

When the FED first acted to increase the interest rate (red vertical line), Bitcoin had <b>the worst performance</b> among all asset classes in this comparison.


This test concludes that bitcoin fails again to meet his safe haven characteristic.

## `Volatility Analysis`



However, let's give it the benefit of the doubt and perform another test. 

Safe heven should at least have a smaller volatility compared to other risky assets. It is pretty clear that bitcoin is way more volatile than the assets in the analysis. However, how big is that volatility? 

```python
for asset in prices_df.columns[0:4]:
   print(f'The annualized volatility for {asset} is currently at: {format(((prices_df[asset] / prices_df[asset].shift()).std() * (252 ** (1/2))) * 100, ".2f")}%')
```

As we can see, the annualized volatility for `BITCOIN` is more than three times higher than the volatility for NASDAQ 100, the riskier asset in this analysis.

While `GOLD`, the actual safe haven, has the lowest annualized volatility in this comparison.
